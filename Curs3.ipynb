## Dualitate

Pentru un spațiu vectorial $L$, spațiul său dual $L^*$ este spațiul funcțiilor liniare $f: L \to \mathbb{K}$. Elementele lui $L^*$ se mai numesc forme liniare sau funcționale liniare. 

**Teoremă** Dacă dimensiunea $L$ este finită atunci $\text{dim }L^*=\text{dim }L$.

Dacă $\mathbf{e}_1,\dots,\mathbf{e}_n$ este o bază a lui $L$. Considerăm $\mathbf{f}_i \in L^*$, definită prin $\mathbf{f}_i(x_1\mathbf{e}_1+\dots+x_n\mathbf{e}_n)=x_i$ pentru orice $i=1,\dots,n$. Arătăm că 
$\mathbf{f}_1,\dots,\mathbf{f}_n$ formează o bază în $L^*.$ Dacă $\mathbf{f}=b_1\mathbf{f}_1+\dots+b_n\mathbf{f}_n$, atunci $\mathbf{f}(\mathbf{x})=b_1x_1+\dots+b_nx_n,$ în particular $\mathbf{f}(\mathbf{e}_j)=b_j$. 
Deci $\mathbf{f}=0$ dacă și numai dacă $\mathbf{f}(\mathbf{x})=0$ pentru orice $\mathbf{x}$, ceea ce înseamnă că $b_1=\dots=b_n$. Am obținut astfel că $\mathbf{f}_1,\dots,\mathbf{f}_n$ sînt liniar-independenți. 
Pe de altă parte orice formă liniară $\mathbf{f}=b_1\mathbf{f}_1+\dots+b_n\mathbf{f}_n$, unde $b_i=\mathbf{f}(\mathbf{e}_i)$. Deci $\text{dim }L^*=\text{dim }L.\Box$

Baza $\mathbf{f}_1,\dots,\mathbf{f}_n$ se numește baza duală a bazei $\mathbf{e}_1,\dots,\mathbf{e}_n$. Se vede că $\mathbf{f}_i(\mathbf{e}_i=1$ și $\mathbf{f}_i(\mathbf{e}_j=0$.

Fie $L'$ un subspațiu al lui $L$. Se numește anulatorul lui $L'$, notat $Ann(L')=\left\{\mathbf{f} \in L^* | f(\mathbf{x})=0,\forall{x} \in L'\right\}$. Să determinăm dimensiunea lui $Ann(L')$. Fie $r=\text{dim } L'$ și
$n=\text{dim } L$. Luăm o bază $\mathbf{e}_1,\dots,\mathbf{e}_r$ a lui $L'$ și o completăm la o bază $\mathbf{e}_1,\dots,\mathbf{e}_n$ a lui $L$. Din definiția bazei duale avem că o funcțională $\mathbf{f} \in Ann(L')$
dacă și numai dacă $\mathbf{f} \in \langle \mathbf{f}_{r+1},\dots,\mathbf{f}_n\rangle$. Deci $Ann(L')=\mathbf{f} \in \langle \mathbf{f}_{r+1},\dots,\mathbf{f}_n\rangle$ și prin urmare $\text{dim}Ann(L')=\text{dim }L - 
\text{dim } L'.$

De asemenea avem că $Ann(L'+L'')=Ann(L') \cap Ann(L'')$.

Dacă $T: L \to M$ este o aplicație liniară atunci definim $T^* M^* \to L^*$ prin $T^*(\mathbf{g})(\mathbf{x})= \mathbf(T(\mathbf{x})$. Luînd două baze arbitrare în $L$ și $M$ și bazele duale asociate atunci matricea lui
$T^*$ este **transpusa** matricei lui $T$.

## Forme biliniare și pătratice

**Definiție** O formă pătratică în $n$ este un polinom omogen de gradul $2$ în variabilele $x_1,\dots,x_n$. Adică este de forma 
$$
\psi(x_1,\dots,x_n)=\sum_{i,j}^n a_{ij}x_ix_j.
$$

Evident putem considera o formă pătratică și ca o funcție de argument vectorial $\mathbf{x}=x_1\mathbf{e}_1+\dots+x_n\mathbf{e}_n$, $\psi(\mathbf{x})=\sum_{i,j}^n a_{ij}x_ix_j.$ Formele pătratice sînt îndeaproape înrudite cu funcțiile liniare. Ca să studiem mai bine legătura considerăm următoarea definiție:

**Definiție** O formă bliniară este o aplicație $b(\mathbf{x},\mathbf{y})$ cu $\mathbf{x,y} \in L$ care este liniară în fiecare argument, adică
$$
\begin{aligned}
\psi(\mathbf{x}_1+\mathbf{x}_2,\mathbf{y}) & = & \psi(\mathbf{x}_1,\mathbf{y})+\psi(\mathbf{x}_2,\mathbf{y}) \\
\psi(a\mathbf{x},\mathbf{y}) & = & a\psi(\mathbf{x},\mathbf{y}) \\
\psi(\mathbf{x},\mathbf{y}_1+\mathbf{y}_2) & = & \psi(\mathbf{x},\mathbf{y}_1)+\psi(\mathbf{x},\mathbf{y}_2) \\
\psi(\mathbf{x},a \mathbf{y}) & = & a\psi(\mathbf{x},\mathbf{y}). 
\end{aligned}
$$

Dacă fixăm o bază a lui $L$, atunci $\psi(\mathbf{x},\mathbf{y})=\sum_{i,j}^n a_{ij}x_iy_j$ cu $a_{ij}=\psi(\mathbf{e}_i,\mathbf{e}_j).$ Matricea care se obține $A=\left(a_{ij}\right)$ se numește matricea coeficienților
lui $\psi$. Se vede de aici că o formă biliniară este complet determinată de matricea coeficienților (de îndată ce am fixat o bază a lui $L$). De asemenea se vede ă dacă înlocuim $\mathbf{y}$ cu $\mathbf{x}$ obținem o 
formă pătratică și mai mult orice formă pătratică se obține în acest fel. 

Este evident că formele biliniare formează un spațiu vectorial. Acesta este izomorf cu spațiul $\mathcal{L}(L,L^*)$ al aplicațiilor liniare $L \to L^*$.

**Definiție** O aplicație biliniară $\psi$ se numește
   - simetrică dacă $\psi(\mathbf{x},\mathbf{y})=\psi(\mathbf{y},\mathbf{x})$;
   - anti-simetrică $\psi(\mathbf{x},\mathbf{y})=-\psi(\mathbf{y},\mathbf{x})$;
   
Se vede că o formă este simetrică dacă matricea coeficienților este simetrică ($A=A^t$) într-o anumită bază și la fel este anti-simetrică dacă matricea coeficienților este anti-simetrică într-o bază.

Să presupunem că $\mathbf{x}=\left(x_1,\dots,x_n\right)$ și $\mathbf{y}=\left(y_1,\dots,y_n\right)$. Putem scrie o formă biliniară prescurtat 
$\psi(\mathbf{x},\mathbf{y})=\mathbf{x}A\mathbf{y}^t=\mathbf{y}A^*\mathbf{x}^t.$  Presupunem că $\mathbf{e}_1,\dots,\mathbf{e}_n$ și $\mathbf{e}_1',\dots,\mathbf{e}_n'$ două baze ale lui $L$ cu matricea de trecere $C$. 
Avem $\psi(\mathbf{x},\mathbf{y})=\mathbf{x}A\mathbf{y}^t=\mathbf{x}'A'\mathbf{y}'^t$. Avem că $A'=C^tAC.$ De aici rezultă că rangul lui $A$ coincide cu rangul lui $A'$. Deci putem defini rangul unei forme biliniare. 
Dacă rangul $r$ al acesteia satisface $r=n$, atunci spunem că $\psi$ este nedegenerată. 

Aceeași formă pătratică poate proveni din mai multe forme biliniare. Aceasta provine din faptul că în expresia unei forme pătratice apar termeni similari. Avem totuși următoarea teoremă

**Teoremă** Pentru orice formă pătratică $\phi$ există și este unică o formă biliniară simetrică $\psi$ astfel încît $\phi(\mathbf{x})=\psi(\mathbf{x},\mathbf{x}).$

**Teoremă** Pentru un corp de caracteristică diferită de $2$, dacă $\psi$ este o formă antisimetrică atunci $\psi(\mathbf{x},\mathbf{x})=0$. Reciproc dacă $\psi(\mathbf{x},\mathbf{x})=0$ pentru orice $\mathbf{x} \in L$ 
atunci $\psi$ este anti-simetrică.

Faptul că orice formă pătratică provine dintr-o formă biliniară simetrică, avem că în expresia $\psi(x_1,\dots,x_n)=\sum_{i,j=1}^n a_{ij}x_ix_j$ $a_{ij}=a_{ji}$, adică termenul corespunzător lui $x_ix_j$ apare de două 
ori, iar în expresia finală va avea coeficientul $2a_{ij}.$

### Forma canonică a unei forme pătratice

Ca și pentru aplicații liniare, forma canonică a unei forme pătratice se obține prin alegerea unei baze speciale în care matricea sa are forma diagonală. Pentru aceasta avem nevoie de noțiunea de ortogonalitate. 
Presupunem de acum că forma $\psi$ este simetrică.

**Definiție** Doi vectori $\mathbf{x,y} \in L$ se numesc ortogonali în raport cu $\psi$ dacă $\psi(\mathbf{x,y})=0$. Spunem că vectorul $\mathbf{x}$ este ortogonal pe un subspațiu $L'$ dacă este ortogonal pe orice vector din $L'.$ 

Mulțimea tuturor vectorilor ortogonali pe un subspațiu dat $L'$ este la rîndul său un subspațiu vectorial notat $(L')^\perp_\psi$ și numit **complementul ortogonal** al lui $L'$ în raport cu $\psi$. Atunci cînd $L'=L$ 
spațiul $L^\perp_\psi$ se numește **radicalul** lui $\psi$. 

Un vector $\mathbf{x}$ aparține radicalului dacă și numai dacă $\psi(\mathbf{x},\mathbf{e}_i)=0$  pentru toți vectorii $\mathbf{e}_i$ dintr-o bază. Dacă descompunem vectorul $\mathbf{x}$ în raport cu această bază obținem un sistem omogen cu matricea coeficienților $A$. Din teorema dimensiunii avem că $\text{dim } L^\perp_\psi=\text{dim }L-r$, unde $r$ este rangul lui 
$\psi$.

**Teoremă** Fie $\psi$ o formă biliniară și $L'$ in subspațiu astfel încît restricția lui $\psi$ la $L'$ este nedegenerată. Atunci $L=L' \oplus (L')^\perp_\psi.$ 

Acest rezultat ne conduce la următorul:

**Teoremă** Pentru orice formă pătratică există  o bază a lui $L$ astfel încît $\psi(\mathbf{x})=\lambda_1x^2_1+\dots+\lambda_nx^2_n.$

Aceasta se numește o formă canonică a lui $\psi$. Această formă nu este unică. Avem că în baza găsită matricea formei biliniare asociate este 
$$
\left(\begin{matrix}\lambda_1 & & & \\
& \lambda_2 & & \\
 &  & \ddots &  \\
& & & \lambda_n\end{matrix}\right).
$$

Numărul de valori nenule este egal cu rangul lui $\psi$ și nu depinde de alegerea bazei. În termeni de matrice rezultatul poate fi reformulat astfel:

**Teoremă** Pentru o matrice simetrică $A$ exită o matrice nesingulară $C$ astfel încît $C^tAC$ este diagonală. Pentru matrice nesingulare diferite matricele diagonale vor fi diferite,
dar numărul de valori nenule de pe diagonală va rămîne neschimbat.

Avem un rezultat similar pentru forme anti-simetrice:

**Teoremă** Fie $\psi$ o formă biliniară anti-simetrică pe spațiul vectorial $L$. Atunci există o bază $\mathbf{e}_1,\dots,\mathbf{e}_n$, astfel că primii $2r$ vectori satisfac 
$\psi(\mathbf{e}_{j-1},\mathbf{e}_j)=1$ și $\psi(\mathbf{e}_i,\mathbf{e}_j)=0$ pentru $i > 2r$, $j > 2r$ sau $|i-j| > 1$. În raport cu aceasta bază matricea lui $\psi$ este
$$
\left(\begin{matrix} 0 & 1 & \dots & \dots & \dots & \dots & \dots & \dots & \dots & 0 \\
-1 & 0 & \dots & \dots & \dots & \dots & \dots & \dots & \dots & 0 \\
\dots & \dots & 0 & 1 & \dots & \dots & \dots & \dots & \dots & 0 \\
\dots & \dots & -1 & 0 & \dots & \dots & \dots & \dots & \dots & 0 \\
 &  & & & \ddots &  &  &  &  &  \\
 \dots & \dots & \dots & \dots & \dots & 0 & 1 & \dots & \dots & 0 \\
\dots & \dots & \dots & \dots & \dots & -1 & 0 & \dots & \dots & 0 \\
\dots & \dots & \dots & \dots & \dots & \dots & \dots & 0 & \dots & 0 \\
&  & & & &  &  & & \ddots  &   \\
0 & \dots & \dots & \dots & \dots & \dots & \dots & \dots & \dots & 0 
\end{matrix}\right).
$$

## Forme reale și complexe

Să presupunem acum că $\phi$ este o formă pătratică peste un spațiu vectorial complex $L$. Am văzut că există o bază $\mathbf{e}_1,\dots,\mathbf{e}_n$ astfel încît $\phi$ are forma 
$\psi(\mathbf{x}=\lambda_1x^2_1+\dots\lambda_nx_n^2.$ Presupunem că $\lambda_1,\lambda_r \neq 0$ și $\lambda_{r+1}=\dots=\lambda_n=0$. Vom lua o nouă bază 
$\mathbf{e}_j'=\sqrt{\lambda_j}\mathbf{e}_j$ pentru $j\leq r$ și $\mathbf{e}'_j=\mathbf{e}_j$ pentru $j > r$. Atunci $\phi(\mathbf{x})=x_1^2+\dots x_r^2.$

În cazul real nu putem proceda la fel de simplu. Va trebui să separăm coeficienții pozitivi $\lambda_1,\dots,\lambda_s$ de cei negativi $\lambda_{s+1},\dots,\lambda_r$. Forma lui $\phi$
ca fi $\phi(\mathbf{x})=x_1^2+\dots+x_r^2-x_{r+1}^2-\dots-x_s^2.$

**Definiție** O formă pătratică reală se numește **pozitiv definită** dacă $\psi(\mathbf{x}) > 0$ pentru orice $\mathbf{x} \neq 0$. Dacă $\psi(\mathbf{x}) < 0$ pentru orice 
$\mathbf{x} \neq 0$, atunci $\psi$ se numește **negativ definită.**

Numărul $s$ care apare în descompunerea lui $\psi$ este dimensiunea celui mai mare subspațiu $L'$ pentru care restricția lui $\psi$ este pozitiv-definită. Acest număr se numește 
**indicele de inerție** al formei $\psi$.

Avem un criteriu care ne permite să decidem dacă o formă este pozitiv definită sau nu. Dacă $\psi$ este o formă pătratică cu matricea asociată $A$, notăm cu $\Delta_i$ minorul format de 
primele $i$ coloane și $i$ linii.

**Teoremă** (criteriul lui Sylvester) O formă pătratică este pozitiv definită dacă și numai dacă $\Delta_i > 0$ pentru orice $i=1,\dots,n$.


## Spații euclidiene

Un spațiu euclidian este un spațiu vectorial real $L$ împreună cu o formă biliniară, simetrică și pozitiv definită. Forma biliniară se numește **produs scalar** și se notează 
$\mathbf{x}\cdot\mathbf{y}.$ 

Pe $\mathbb{R}^n$ se poate defini produsul scalar canonic: $\mathbf{x}\cdot\mathbf{y}=x_1y_1+\dots+x_ny_n.$  Lungimea unui vector se definește 
$|\mathbf{x}|=\sqrt{\mathbf{x}\cdot\mathbf{x}}$. De asemenea definim unghiul dintre doi vectori:
$$
\cos \phi = \frac{\mathbf{x}\cdot\mathbf{y}}{|\mathbf{x}|\cdot|\mathbf{y}|}.
$$

Dacă fixăm un vector $\mathbf{e}\neq 0$ orice vector $\mathbf{x} \in L$ poate fi scris $\mathbf{x}=\alpha\mathbf{e}+\mathbf{y}$ cu $\mathbf{e}\cdot\mathbf{y}=0$. Vectorul 
$\alpha\mathbf{e}$ se numește proiecția ortogonală pe dreapta $\langle\mathbf{e}\rangle.$ Lungimea proiecției ortogonale este cel mult $|\mathbf{x}|.$

**Teoremă** Pentru orice vectori $\mathbf{x},\mathbf{y} \in L$, $|\mathbf{x}\cdot\mathbf{y}| \leq |\mathbf{x}|\cdot|\mathbf{y}|.$

Din această teoremă rezultă că $|x+y| \leq |x|+|y|.$

**Definiție** O bază a $\mathbf{e}_1,\dots,\mathbf{e}_n$ a unui spațiu euclidian se numește ortonormată dacă $|\mathbf{e}_i|=1$ pentru orice $i$ și $\mathbf{e}_i\cdot\mathbf{e}_j=0$ 
pentru $i\neq j$.

**Teoremă** Orice spațiu euclidian are o bază ortonormată.

### Aplicații ortogonale

**Definiție** O aplicație $T:L \to L$ a unui spațiu euclidian se numește ortogonală dacă $T(\mathbf{x})\cdot T(\mathbf{y})=\mathbf{x}\cdot\mathbf{y}$.

Dacă fixăm o bază ortonormată a lui $L$ și $T$ este o aplicație ortogonală avînd matricea $U$, atunci $U^tU=UU^t=I_n$, unde $n=\text{dim } L.$

O aplicație $T:L \to L$ se numește simetrică, dacă $(L\mathbf{x})\cdot\mathbf{y}=\mathbf{x}\cdot L\mathbf{y}$. Într-o bază ortonormată matricea lui $T$ va 
fi simetrică. Privind structura acestui tip de aplicații avem următorul rezultat:

**Teoremă** Pentru orice aplicație simetrică $T$ există o bază ortonormată $\mathbf{e}_1,\dots,\mathbf{e}_n$ a lui $L$ formată din vectori proprii ai lui
$T$.

**Teoremă** Dacă $T$ este aplicație simetrică, atunci la valori proprii distincte corespund vectori proprii ortogonali.

Revenind la forme pătrattice avem următoarea aplicație:

**Teoremă** Pentru orice formă pătratică există o bază ortonormată în raport cu care $\psi(\mathbf{x})=\lambda_1x_1^2+\dots\lambda_nx_n^2$. Coeficienții
$\lambda_1,\dots,\lambda_n$ sînt unici pînă la o permutare.

### Suprafețe de gradul $2$

Vom numi suprafață de gradul 2 în spațiul $\mathbb{R}^n$ mulțimea punctelor $\mathbf{x}=(x_1,\dots,x_n)$ care satisfac o  ecuație de forma
$$
\sum_{i,j=1}^n a_{ij}x_ix_j+\sum_{i=1}^n b_ix_i+c=0 
$$
sa pe scurt $\psi(\mathbf{x})+L(\mathbf{x})+c=0$, unde $\psi$ este o formă pătratică, $L$ o formă liniară și $c \in \mathbb{R}$. Cînd $n=2$ numim o astfel
de suprafață (de fapt o curbă) conică, iar pentru $n=3$ o numim cuadrică.

În mod practic procedăm în felul următor: Mai întîi facem o schimbare ortogonală de coordonate $x_i=\sum_{j=1}^n q_{ij}y_j$ astfel că 
$\psi(\mathbf{x})=\lambda_1y_1^2+\dots+\lambda_ny_n^2$. Deci ecuația suprafeței devine
$$
\sum_{j=1}^n\lambda_jy_j^2+\sum_{j=1}^n l_jy_j +c=0.
$$

Dacă un coeficient $\lambda_j\neq 0$, atunci putem scrie 
$$
\lambda_jy_j^2+l_jy_j=\lambda_j\left(y_j+\frac{l_j}{\lambda_j}\right)^2-\frac{l_j^2}{\lambda_j}.
$$
Dacă punem $z_j=\left(y_j+\frac{l_j}{\lambda_j}\right)$ termenul corespunzător de gradul $2$ rămîne înmulțit cu $\lambda_j$, termenul de gradul $1$ 
dispare, iar la termenul liber se adaugă $-\frac{l_j^2}{\lambda_j}$. 

Ecuația suprafeței va deveni
$$
\lambda_1z_1^2+\dots\lambda_rz_r^2+l_{r+1}z_{r+1}+\dots l_nz_n+c'=0.
$$

 - Dacă $r=n$ sau $l_j=0$ pentru orice $r+1 \leq j \leq n$, atunci obținem o **suprafață cu centru**. În primul caz se numește proprie, dacă $c \neq 0$,iar
   altfel se numește conică.
 - Dacă există un $l_k \neq 0$ putem face o nouă schimbare ortogonală de coordonate 
   $$
   \begin{aligned}
   t_1 & = z_1 \\
   \vdots \\
   t_{r+1} & = \left(-l_{r+1}z_{r+1}+\dots+l_nz_n\right)\frac{1}{M}\\
   \vdots
   \end{aligned}
   $$
   cu $M$ ales astfel încît aplicația să fie ortogonală, adică
   $$
   M^2=l_{r+1}^2+\dots+l_n^2.
   $$
   Ecuația devine $\lambda_1t_1^2+\dots+\lambda_r t_r^2=2Mt_{r+1}-c$. Făcînd o ultimă translație obținem $\lambda_1t_1^2+\dots+\lambda_r t_r^2=2Mt_{r+1}.$
   Aceasta este ecuația canonică unei suprafețe fără centru.

## Exemple:

Să luăm acum o suprafață cu centru $\lambda_1z_1^2+\dots\lambda_nz_n^2=c$, cu $c \neq 0$. Ecuația poate fi pusă sub forma
$$
\frac{z_1^2}{a_1^2}+\dots\frac{z_k^2}{a_k^2}-\frac{z_{k+1}^2}{a_{k+1}^2}-\frac{z_n^2}{a_n^2}=1.
$$

Cînd $n=2$ avem două posibilități
 - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}=1$ Elipsa (cînd $a_1=a_2$ obținem un cerc);
 - $\frac{z_1^2}{a_1^2}-\frac{z_2^2}{a_2^2}=1$ Hiperbola.
Pentru $n=3$ avem 
  - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}+\frac{z_3^2}{a_3^2}=1$ Elipsoidul (cînd $a_1=a_2=a_3$ obținem o sferă);
  - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}-\frac{z_3^2}{a_3^2}=1$ Hiperboloidul cu o pînză;
  - $\frac{z_1^2}{a_1^2}-\frac{z_2^2}{a_2^2}-\frac{z_3^2}{a_3^2}=1$ Hiperboloidul cu două pînze.
  
Cazul conic:

Pentru $n=2$ avem 
  - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}=0$ Un punct;
  - $\frac{z_1^2}{a_1^2}-\frac{z_2^2}{a_2^2}=0$ O pereche de drepte.
În cazul $n=3$ avem 
   - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}+\frac{z_3^2}{a_3^2}=0$ un punct;
   - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}-\frac{z_3^2}{a_3^2}=0$ un con eliptic;
   - $\frac{z_1^2}{a_1^2}-\frac{z_2^2}{a_2^2}-\frac{z_3^2}{a_3^2}=0$ un con hiperbolic;

Suprafețe fără centru:

Pentru $n=2$ avem:
 - $z_1^2=2a_1^2 z_2$ parabola;
 - $z_1^2=0$ o dreaptă dublă;

Pentru $n=3$ avem:
   - $\frac{z_1^2}{a_1^2}+\frac{z_2^2}{a_2^2}=2z_3$ Paraboloid eliptic;
   - $\frac{z_1^2}{a_1^2}-\frac{z_2^2}{a_2^2}=2z_3$ Paraboloid hiperbolic;
Dacă în ecuația unei cuadrice nu apare deloc una dintre variabile, adică formal avem ecuația unei conice atunci obținem cilindrii: eliptic, 
hiperbolic,parabolic sau o intersecție de plane.
